```python
import dirAnalysis

da = dirAnalysis.DirAnalysis()
```

```python
# Добавление данных в бд из паки Music и только файлы mp3
da.set_data("Music", "audiofile.db",type_file=[".mp3"])
da.output_name_data_db("audiofile.db")
```

```python
# Поиск в базе данных по названию файла
data = da.get_search("audiofile.db", "name", "Hometown - twenty one pilots.mp3")
```

```python
# Поиск в базе данных по названию песни
data = da.get_search("audiofile.db", "title", "Monster")
```

```python
# Поиск в базе данных по году
data = da.get_search("audiofile.db", "year", "2013")
```

```python
# Фильтрация данных поиска по году, филтрация происходит по имени файла и по алфавиту: а-я
data = da.get_filter("audiofile.db", "name", True)
```

```python
# Поиск в базе данных по артисту
data = da.get_search("audiofile.db", "artist", "Imagine Dragons")
```

```python
# Фильтрация данных поиска по артисту, филтрация происходит по имени файла и по алфавиту: я-а
data = da.get_filter("audiofile.db", "name", False)
```

```python
# Поиск в базе данных по жанру
data = da.get_search("audiofile.db", "genre", "Alternative Rock")
```

```python
# Фильтрация данных поиска по жанру, филтрация происходит по году и по возраствнию
data = da.get_filter("audiofile.db", "year", True)
```

```python
# Фильтрация данных поиска по жанру, филтрация происходит по году и по убыванию
data = da.get_filter("audiofile.db", "year", False)
```

```python
# Сбросить данные поиска
da.cancel_search()
```
